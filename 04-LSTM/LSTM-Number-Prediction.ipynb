```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate a simple sequence dataset (e.g., predicting next number in a sequence)
sequence_length = 5  # Use past 5 numbers to predict the next one

# Generate training data (sequences like [1,2,3,4,5] → 6, [2,3,4,5,6] → 7)
data = np.array([i for i in range(1, 101)])  # Numbers 1 to 100

X, y = [], []
for i in range(len(data) - sequence_length):
    X.append(data[i:i + sequence_length])
    y.append(data[i + sequence_length])

X, y = np.array(X), np.array(y)

# Reshape X for LSTM (samples, time steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build the LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(sequence_length, 1)),  # LSTM layer
    Dense(1)  # Output layer (predicts next number)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=200, verbose=1)

# Function to predict the next number
def predict_next(sequence):
    sequence = np.array(sequence).reshape(1, sequence_length, 1)  # Reshape input
    return model.predict(sequence)[0][0]

# Try predicting the next number in a sequence
test_sequence = [96, 97, 98, 99, 100]  # Predict what comes after 100
predicted_number = predict_next(test_sequence)
print(f"Predicted next number after {test_sequence}: {predicted_number:.2f}")
```
